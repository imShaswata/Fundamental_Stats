    <br> = $\frac{1}{6}$ for $ x= 1,2,3,4,5,6$

        - Variance : 
        <br> V(X) = $E [ X-E(X) ]^2 = E(X^2) - E(X)^2$ = $\sum_{i=1}^{6} i^2 P(X=i) - 3.5^2$ = $ \frac{1^2+2^2+..6^2}{6} -3.5^2 $
        
        
- **Bernouli Distribution :**
    - If a RV has only **TWO discrete points** as its range of admissible values called SUCCES and FAILURE (or 0,1) with p being Probability of success

    - PMF : If $X \sim Ber(p) $ then p.m.f of X is given by <br> $P(X=x) = p^x (1-p)^{(1-x)}$ for $x=0,1$

## B. Continous Distributions :
### **Normal/Gaussian Distribution :**

#### Theory


- If a RV has only real number as its range of admissible values with the following form of PDF <br>
$ f_X(x) = \frac{1}{\sqrt{(\sigma^2 2 \pi)}} e- {\frac{1}{2} \frac{(x-\mu)}{\sigma} }^2$ 
with parameters $-\infty < \mu < \infty $ , $\sigma >0 $  and $-\infty < x < \infty $ <br>

- Notation:  $X \sim N(\mu,\sigma^2) $ <br> 

- Moments of Normal Distribution : 
    Expectation : $\mu $ 
    Variance : $\sigma^2$:
    
- **Standard Normal Variable(SNV):** If $\mu = 0$ and  $\sigma= 1$ then it's SNV with <br>
    - pdf :$ f_X(x) = \frac{1}{\sqrt{ 2 \pi}} e- { \frac{x}{2}  }^2$
    - DF  :$ F_X(x) = P[X\leq x] = \int_{-\infty}^{x} \frac{1}{\sqrt{ 2 \pi}} e- { \frac{t}{2}}^2 dt$ 
        - $F_X(x)$ is written particularly as $\Phi(x)$
        - What is $\Phi(1.96)$ ? <br>
        Ans: $\Phi(1.96) = P[X \leq 1.96] = \int_{-\infty}^{1.96} \frac{1}{\sqrt{ 2 \pi}} e- { \frac{t}{2}}^2 dt = 0.975$
        <br> It means : If X is a SNV then 97.5% times X has value less than 1.96 
        
- **Z score / Z value:** Z value is the value of a SNV.
    - $Z_\alpha $ is that point of a SNV 
        - below which $100(1- \alpha) $% of all the value lies OR
        - you have $\alpha$ % chance of observing any value above that point
    - It is used to 
        - determine the position of a raw score in terms of its distance from the mean ,when measured in standard deviation units
        - check the 'abnormality' of your observed result by comparing it to 'ideal' result from a SNV

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss

plt.style.use('seaborn') # pretty matplotlib plots
plt.rcParams['figure.figsize'] = (12, 5)



def plot_gaussisan(x,mu,sigma,q,p=None):
    y_pdf = ss.norm.pdf(x, mu, sigma) # the normal pdf
    y_cdf = ss.norm.cdf(x, mu, sigma) # the normal cdf
    
    if p==None:    
        print("Phi(q) = P[X<=q]=",ss.norm.cdf(q))
        plt.axvline(q,0,1,color='y')
    else:
        print("Phi inv(p) =",ss.norm.ppf(p))
        plt.axvline(p,0,1,color='y')
    plt.plot(x, y_pdf, label='pdf')
    plt.plot(x, y_cdf, label='cdf')
    plt.legend()
```

```python
plot_gaussisan(x = np.linspace(-10, 10, 10000),  mu = 0,sigma = 1,p=0.975,q=None)
```
